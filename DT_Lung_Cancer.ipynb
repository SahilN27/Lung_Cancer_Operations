```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, precision_score
```

```python
Data = pd.read_csv(r'survey lung cancer.csv')
LungCancer_df = pd.DataFrame(Data)
LungCancer_df
```

```python
oe = OrdinalEncoder()
LungCancer_df['GENDER']=oe.fit_transform(LungCancer_df[['GENDER']])
LungCancer_df['AGE']=oe.fit_transform(LungCancer_df[['AGE']])
LungCancer_df['SMOKING']=oe.fit_transform(LungCancer_df[['SMOKING']])
LungCancer_df['YELLOW_FINGERS']=oe.fit_transform(LungCancer_df[['YELLOW_FINGERS']])
LungCancer_df['ANXIETY']=oe.fit_transform(LungCancer_df[['ANXIETY']])
LungCancer_df['PEER_PRESSURE']=oe.fit_transform(LungCancer_df[['PEER_PRESSURE']])
LungCancer_df['CHRONIC DISEASE']=oe.fit_transform(LungCancer_df[['CHRONIC DISEASE']])
LungCancer_df['FATIGUE ']=oe.fit_transform(LungCancer_df[['FATIGUE ']])
LungCancer_df['ALLERGY ']=oe.fit_transform(LungCancer_df[['ALLERGY ']])
LungCancer_df['WHEEZING']=oe.fit_transform(LungCancer_df[['WHEEZING']])
LungCancer_df['ALCOHOL CONSUMING']=oe.fit_transform(LungCancer_df[['ALCOHOL CONSUMING']])
LungCancer_df['COUGHING']=oe.fit_transform(LungCancer_df[['COUGHING']])
LungCancer_df['SHORTNESS OF BREATH']=oe.fit_transform(LungCancer_df[['SHORTNESS OF BREATH']])
LungCancer_df['SWALLOWING DIFFICULTY']=oe.fit_transform(LungCancer_df[['SWALLOWING DIFFICULTY']])
LungCancer_df['CHEST PAIN']=oe.fit_transform(LungCancer_df[['CHEST PAIN']])
LungCancer_df['LUNG_CANCER']=oe.fit_transform(LungCancer_df[['LUNG_CANCER']])
LungCancer_df

```

```python
X= LungCancer_df.iloc[:,0:-1]
y= LungCancer_df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,random_state=1)
model=DecisionTreeClassifier(criterion='gini',random_state=1)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
f1 = f1_score(y_test, y_pred, average='weighted')
print(f"The F1 Score of the Decision Tree Model is {f1}")
```

```python
precision_value = precision_score(y_test, y_pred, average='weighted')
print(f"The Precision Value of the Decision Tree Model is {precision_value}")
```

```python
accuracy= accuracy_score(y_pred,y_test)
print(f"The accuracy of Decision Tree Model is {accuracy*100}%")
```
